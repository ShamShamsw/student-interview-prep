# Study Session Tracker Walkthrough

Build a Python CLI that logs study sessions and summarizes minutes by topic.

## 1) Project goal

Implement commands:

- `list`
- `add <topic> <minutes>`
- `remove <id>`
- `summary`
- `clear`

## 2) Start in `starter/`

```bash
python app.py list
```

## 3) Implement persistence

In `starter/app.py`, complete:

- `load_sessions`
- `save_sessions`

Checklist:
- Return `[]` if file is missing.
- Use UTF-8 JSON read/write.
- Convert dataclass values with `asdict`.

## 4) Implement core operations

Then complete:

1. `add_session`
2. `remove_session`
3. `clear_sessions`
4. `summarize_minutes`

Verification sequence:

```bash
python app.py add "arrays" 45
python app.py add "graphs" 30
python app.py list
python app.py summary
python app.py remove 1
python app.py clear
```

## 5) Compare with final

Compare with `final/app.py` for structure and behavior consistency.

## 6) Extension ideas

- add session date/time
- add CSV export
- add `top-topic`
- add topic filtering
