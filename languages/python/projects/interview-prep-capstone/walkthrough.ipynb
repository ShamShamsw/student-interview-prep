# Interview Prep Capstone Walkthrough (8-12 Weeks)

This walkthrough helps you execute the capstone with architecture-first thinking and Agile delivery.

## 0) Read first (mandatory)

1. `docs/01-design-and-architecture.md`
2. `docs/02-sprint-plan-agile.md`
3. `docs/03-ai-and-resource-playbook.md`

## 1) Sprint setup

- Choose 8-week or 12-week path.
- Create sprint board columns: Ready, In Progress, Done.
- Define your Definition of Done for each ticket.

## 2) Build sequence

### Sprint 1
- Health endpoint and project structure
- Attempt model and creation endpoint
- Baseline tests and CI checks

### Sprint 2
- Progress aggregation endpoint
- Input validation and error envelope consistency

### Sprint 3
- Recommendation logic (weak-topic prioritization)
- Deterministic tests for recommendation ranking

### Sprint 4+
- Study plan endpoint and persistence improvements
- Refactor + hardening + postmortem

## 3) AI usage checkpoints

At each blocker:
- write failing test first
- ask AI with exact error + context
- verify against official docs
- keep smallest safe diff

## 4) Weekly review prompts

- What architecture decision changed this week, and why?
- Which endpoint caused the most defects?
- Which test prevented a regression?
- What will you cut next sprint if schedule slips?

## 5) Final capstone closeout

- Demonstrate API flow end-to-end
- Show sprint board outcomes
- Present architecture tradeoffs
- Deliver retrospective and next roadmap
