# Interview Prep Capstone Walkthrough (8-12 Weeks)

This walkthrough helps execute the capstone with architecture-first thinking and Agile delivery.

## 0) Read first (mandatory)

1. `docs/01-design-and-architecture.md`
2. `docs/02-sprint-plan-agile.md`
3. `docs/03-ai-and-resource-playbook.md`

## 1) Sprint setup

- choose 8-week or 12-week path
- create board columns: Ready, In Progress, Done
- define Definition of Done for each ticket

## 2) Build sequence

### Sprint 1
- health endpoint and project structure
- attempt model and create endpoint
- baseline tests and CI

### Sprint 2
- progress aggregation endpoint
- validation and consistent errors

### Sprint 3
- recommendation logic
- deterministic tests for ranking

### Sprint 4+
- study plan behavior and persistence improvements
- hardening and retrospective

## 3) AI usage checkpoints

At each blocker:
- write failing test first
- ask AI with exact error + context
- verify against official docs
- keep smallest safe diff

## 4) Weekly review prompts

- what architecture decision changed and why?
- which endpoint caused most defects?
- which test prevented a regression?
- what will you cut if schedule slips?

## 5) Final closeout

- demo API end-to-end
- show sprint outcomes
- present architecture tradeoffs
- deliver retrospective and next roadmap
