# Sample TODO App Walkthrough

This walkthrough guides you through building the starter CLI TODO app and validating against the final solution.

## 1) Goal

Build a command-line TODO app with persistent storage (`todos.json`) and these commands:

- `list`
- `add "task"`
- `done <id>`
- `remove <id>`
- `clear`

## 2) Start in the starter folder

Use a terminal in `sample-to-do-app/starter/`:

```bash
python app.py list
```

## 3) Implement persistence first

In `starter/app.py`, implement:

- `load_todos`
- `save_todos`

## 4) Implement item operations

Implement in this order:

1. `add_todo`
2. `mark_done`
3. `remove_todo`
4. `clear_all`

Suggested check sequence:

```bash
python app.py add "Buy groceries"
python app.py add "Practice arrays"
python app.py list
python app.py done 1
python app.py list
python app.py remove 2
python app.py clear
```

## 5) Compare with final

Compare your starter implementation against `final/app.py` focusing on:

- load/save approach
- update-and-persist pattern
- not-found behavior
- output formatting

## 6) Reflection prompts

- Which function was hardest and why?
- What edge case would you add next?
- What feature would you add after MVP?
